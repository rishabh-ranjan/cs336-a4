```python
import random
from cs336_data.gopher import gopher_quality_filter
from tqdm.auto import tqdm
from IPython.display import clear_output
```

```python
quality = []
for i in tqdm(range(1, 48369)):
    with open(f"../data/warc_txt/{i}.txt", "r") as f:
        text = f.read()
    quality.append(gopher_quality_filter(text))
```

```python
while True:
    sample = random.randint(1, 48369)
    with open(f"../data/warc_txt/{sample}.txt", "r") as f:
        text = f.read()
    qual = quality[sample]
    print(text)
    print("---")
    print(qual)
    inp = input()
    if inp == "q":
        break
    clear_output(wait=True)
```
