```python
import pandas as pd
import torch
from cs336_data.mask import mask_emails, mask_ips, mask_phone_numbers
```

```python
pd.set_option('display.max_colwidth', None)
```

```python
perm = torch.randperm(48_368)
perm += 1
```

```python
count = 0
with open("../data/unmasked.txt", "w") as unmasked_f, open("../data/masked.txt", "w") as masked_f:
    for sample in perm:
        with open(f"../data/warc_txt/{sample}.txt", "r") as f:
            text = f.read()
        masked_text, _ = mask_emails(text)
        masked_text, _ = mask_phone_numbers(masked_text)
        masked_text, _ = mask_ips(masked_text)
        if masked_text == text:
            continue
        count += 1
        unmasked_f.write(f"\n=== [unmasked] sample {count}: document {sample} ===\n")
        masked_f.write(f"\n=== [masked] sample {count}: document {sample} ===\n")
        unmasked_f.write(text)
        masked_f.write(masked_text)
        if count == 20:
            break
```
